```python
def calculate_outliers(df_in, col_name, lower_quantile=0.25, upper_quantile=0.75):
    """
    Calculates the number of outliers in a given column of a DataFrame based on the IQR method.

    Parameters:
        df_in (pd.DataFrame): Input DataFrame.
        col_name (str): Name of the column to check for outliers.
        lower_quantile (float): Lower quantile for IQR calculation. Default is 0.25.
        upper_quantile (float): Upper quantile for IQR calculation. Default is 0.75.

    Returns:
        dict: A dictionary containing the number of outliers and the indexes of these outliers.
    """
    q1 = df_in[col_name].quantile(lower_quantile)
    q3 = df_in[col_name].quantile(upper_quantile)
    iqr = q3 - q1  # Interquartile range
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    outliers_df = df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)]

    return {"Number of Outliers": len(outliers_df), "Indexes": outliers_df.index.tolist()}
```

```python
for column_name in ["Recency", "Frequency", "Monetary"]:
    # Call the updated function and store its return value
    outliers_info = calculate_outliers(RFM_df, column_name)

    # Print the results in a more structured format
    print(f"Column: {column_name}")
    print(f"Number of Outliers: {outliers_info['Number of Outliers']}")
    print(f"Indexes of Outliers: {outliers_info['Indexes']}")
    print("*" * 40)
```

```python
def remove_outliers(df_in, col_name):
    """
    Removes outliers from a specified column in a DataFrame based on the IQR method.

    Parameters:
        df_in (pd.DataFrame): The input DataFrame.
        col_name (str): The name of the column from which to remove outliers.

    Returns:
        pd.DataFrame: A new DataFrame with outliers removed from the specified column.
    """
    q1 = df_in[col_name].quantile(0.05)
    q3 = df_in[col_name].quantile(0.95)
    iqr = q3 - q1
    fence_low  = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr

    # Identify indexes of outliers
    index_outliers = df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)].index

    # Drop outliers and return the modified DataFrame directly
    df_out = df_in.drop(index=index_outliers)

    # It's usually not a good practice to print from within a function like this,
    # but to follow your initial pattern:
    print(f"Outliers in the {col_name} column have been removed.")
    return df_out
```

```python
# Clean up city names and payment types for better consistency and readability
clvt_df['customer_city'] = clvt_df['customer_city'].str.title()  # Format city names to title case
clvt_df['payment_type'] = clvt_df['payment_type'].str.replace('_', ' ').str.title()  # Replace underscores with spaces and format to title case in payment types

# Engineering new columns to provide more insights into delivery times and order timestamps
# Calculate the difference in days between the estimated delivery date and the actual delivery date
clvt_df['delivery_against_estimated'] = (clvt_df['order_estimated_delivery_date'] - clvt_df['order_delivered_customer_date']).dt.days

# Extract year, month, day of the week, and hour from the order purchase timestamp for detailed analysis
clvt_df['order_purchase_year'] = clvt_df['order_purchase_timestamp'].dt.year
clvt_df['order_purchase_month'] = clvt_df['order_purchase_timestamp'].dt.month
clvt_df['order_purchase_dayofweek'] = clvt_df['order_purchase_timestamp'].dt.dayofweek
clvt_df['order_purchase_hour'] = clvt_df['order_purchase_timestamp'].dt.hour

# Map numerical day of the week to actual day names for readability
day_of_week_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
clvt_df['order_purchase_day'] = clvt_df['order_purchase_dayofweek'].map(day_of_week_map)

# Map numerical month to month names for readability and standardize month representation
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
clvt_df['order_purchase_mon'] = clvt_df['order_purchase_month'].map(month_map)

# Standardize the month-year format for consistent analysis and sorting
clvt_df['month_year'] = clvt_df['order_purchase_year'].astype(str) + '-' + clvt_df['order_purchase_month'].apply(lambda x: f'{x:02d}')

# Create a 'year-month' column for temporal analysis, ensuring proper sorting by using numerical year and month concatenation
clvt_df['month_y'] = clvt_df['order_purchase_timestamp'].apply(lambda date: 100 * date.year + date.month)
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.stattools import adfuller
```

```python
df_forcast['month_year'] = pd.to_datetime(df_forcast['month_year'])
df_forcast.set_index('month_year', inplace=True)
```

```python
# ADF test function
def adf_test(series, name=''):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Statistic for {name}: {result[0]}')
    print(f'p-value for {name}: {result[1]}')
    if result[1] <= 0.05:
        print(f"{name} is stationary")
    else:
        print(f"{name} is not stationary")

# Perform ADF test on each series
for column in df_forcast.columns:
    adf_test(df_forcast[column], name=column)

# Differencing the series
df_forcast_differenced = df_forcast.diff().dropna()

# Recheck stationarity after differencing
for column in df_forcast_differenced.columns:
    adf_test(df_forcast_differenced[column], name=column)

# Fit the VAR model on the differenced data
model_updated = VAR(df_forcast_differenced)
model_updated_fitted = model_updated.fit()

# Forecasting the next 3 months
forecast_steps = 3
forecast_updated = model_updated_fitted.forecast(df_forcast_differenced.values[-model_updated_fitted.k_ar:], steps=forecast_steps)

# Create a DataFrame for the forecast
df_forecast_updated = pd.DataFrame(forecast_updated, index=pd.date_range(start=df_forcast_differenced.index[-1], periods=forecast_steps+1, closed='right'), columns=df_forcast_differenced.columns)

print(df_forecast_updated)
```

```python

# Assuming df_forecast_updated and df_updated are already defined

# Get the last date from the original dataframe
last_date = df_forcast.index[-1]

# Generate a new index for the forecasted values, ensuring monthly frequency
forecast_index = pd.date_range(start=last_date + pd.offsets.MonthBegin(1), periods=forecast_steps, freq='MS')

# Update the forecast dataframe's index to reflect the correct forecast months
df_forecast_updated.index = forecast_index

# Reverse the differencing to get the forecasted values in the original scale
last_values = df_forcast.iloc[-1]  # Last known values from the original, undifferenced data

# Initialize a dataframe to hold the reversed forecasts
reversed_forecast_df = pd.DataFrame(index=df_forecast_updated.index, columns=df_forecast_updated.columns)

for column in df_forecast_updated.columns:
    # The first forecasted value is added to the last known value from the original data
    reversed_forecast_df[column].iloc[0] = last_values[column] + df_forecast_updated[column].iloc[0]

    # Subsequent forecasted values are cumulatively added to the previous forecasted value
    for i in range(1, len(df_forecast_updated)):
        reversed_forecast_df[column].iloc[i] = reversed_forecast_df[column].iloc[i-1] + df_forecast_updated[column].iloc[i]

# Convert the forecasted values back to their original scale
reversed_forecast_df = reversed_forecast_df.astype(float)

# Append the reversed forecasted values to the original dataframe
final_df_with_months = pd.concat([df_forcast, reversed_forecast_df])

print(final_df_with_months)
```
